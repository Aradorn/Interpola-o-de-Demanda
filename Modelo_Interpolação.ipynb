<h3>Funções Criadas

```python
#Classe para estatística descritiva

class estatistica_descritiva():
    
    def __init__(self,dados):
        self.base = dados
    
    def media(self):
        return round(self.base.mean(),ndigits=2)
    
    def mediana(self):
        return round(self.base.median(),ndigits=2)
    
    def moda(self):
        return round(self.base.mode(),ndigits=2)
    
    def desv_pad(self):
        return round(self.base.std(),ndigits=2)
    
    def quantil(self):
        return round(self.base.quantile(),ndigits=2)
    
    def maximo(self):
        return round(self.base.max(),ndigits=2)
    
    def minimo(self):
        return round(self.base.min(),ndigits=2)
    
    def quartis(self):
        q75, q25 = np.percentile(self.base,[75,25])
        return q75, q25
    
    def norm_test_shapiro(self):
        return scipy.stats.shapiro(self.base)
    
    def norm_test_anderson(self):
        return scipy.stats.anderson(self.base)
    
    def estatistica_geral(self):
        print("Média: ",self.media(),"\n")

        print("Mediana: ",self.mediana(),"\n")
        print("Moda:\n",self.moda(),"\n")

        print("Quantil: ",self.quantil(),"\n")
       
        print("Desvio Padrão: ",self.desv_pad(),"\n")

        print("Amplitude: \n","Máximo: ",self.maximo(),"\n","Mínimo: ",self.minimo(),"\n")

        print("Quartil: \n", "1st: ",self.quartis()[0],"\n","3rd: ",self.quartis()[1])

        #Teste de Normalidade Shapiro-Wilk
        print("Teste de Shapiro-Wilk: ",self.norm_test_shapiro())
        #Teste de Normalidade Anderson Darling
        #print("Teste de Anderson Darling: ",self.norm_test_anderson())
    
    
```

```python
#Classe para decomposição gráfica dos dados
class decompose():
    def __init__(self,dados):
        self.base = dados
        self.decomposicao = seasonal_decompose(self.base, freq=30, extrapolate_trend='freq')
        self.tendencia = self.decomposicao.trend
        self.sazonal = self.decomposicao.seasonal
        self.aleatorio = self.decomposicao.resid
        plt.rcParams['figure.figsize'] = (18,5)
                                          
    def graf_original(self):
        plt.plot(self.base)
        plt.title("Original")
        plt.show()
        
    def graf_tendencia(self):
        plt.plot(self.tendencia)
        plt.title("Tendência")
        plt.show()

    def graf_sazonal(self):
        plt.plot(self.sazonal)
        plt.title("Sazonal")
        plt.show()
        
    def graf_aleatorio(self):
        plt.plot(self.aleatorio)
        plt.title("Aleatorio")
        plt.show()
    
    def all_graphics(self):
        self.graf_original()
        self.graf_tendencia()
        self.graf_sazonal()
        self.graf_aleatorio()
```

<h3><b>Conclusões</b></h3>
<p>Os dados de Série Temportal não apresentam normalidade, portanto, os dados de estatística descritiva não podem ser utilizados para conclusões de distribuição do modelo, sendo então necessário a limpeza e tratamento para normalização para posterior utilização nos modelos de estimativa.
<p>A Correlação entre a temperatura e a venda é apresentada é de 31,96%, ou seja, segundo <i>'Silvia E Shimakura 2006-08-30'</i> a correlação é fraca e positiva onde quanto maior a temperatura, maior a demanda.
<p>Conforme demonstrado pelo Boxplot, existem muitos outliers nos dados que justificam a fuga de normalidade apresentada nos dados. Para normalização dos dados, os Outliers serão removidos para nova análise de normalidade.
Como as análises e posterior estimativas serão sobre cada registro de temperatura individualmente, nesta etapa tanto a remoção de Outliers quanto a análise de normalidade, serão realizadas individualmente sobre cada temperatura registrada, a fim de que o modelo não tenha influência da variação aleatória de temperatura sobre às estatísticas.

```python
#Remoção de Outliers fora do 2º e 3º Quartil

dados_total = dados_total[(dados_total.temperatura_range >= 16)&(dados_total.temperatura_range <= 27)]
dados_sem_outliers = dados_total[(dados_total.temperatura_range == 0)]
registros_temperatura = dados_total.groupby('temperatura_range').count()
for temperatura in registros_temperatura.index:
    outliers = dados_total[(dados_total.temperatura_range == temperatura)]
    quartil = estatistica_descritiva(outliers["Total"]).quartis()
    outliers = outliers[(outliers.Total < quartil[0])&(outliers.Total > quartil[1])]
    dados_sem_outliers = dados_sem_outliers.append(outliers)

registros_sem_outliers = dados_sem_outliers.groupby('temperatura_range').count()
```

```python
#Teste de Normalidade para cada Range de Temperatura
for temperatura in registros_sem_outliers.index:
    teste_norm = dados_sem_outliers[(dados_sem_outliers.temperatura_range == temperatura)]
    print(temperatura, estatistica_descritiva(teste_norm.Total).norm_test_shapiro())
```

<h3><b>Aplicação de Previsões em Valores Tratados</b></h3>

```python
class previsores():
    def __init__(self,dados):
        self.base = dados
        self.model_interpl = sp.interpolate.interp1d(self.base.Temperatura,self.base.Total)
        
    def interpolacao(self):
        return self.model_interpl
    def prev_interpl(self,x):
        yipl = self.model_interpl(x)
        return yipl
    
    def linear_regression(self):
        xlm = np.array(self.base.Temperatura).reshape(-1, 1)
        lnmodel = linear_model.LinearRegression().fit(xlm,self.base.Total)
        return lnmodel

    def prev_linear(self,x):
        ylm = lnmodel.predict(np.array(x).reshape(-1, 1))
        return ylm
    
```

```python
#Filtro de dados com mais de 5% de representatividade

dados_filtrados = dados_sem_outliers[(dados_sem_outliers.temperatura_range >= 16)&(dados_sem_outliers.temperatura_range <= 27)]
```

```python
class treino_teste():
    def __init__(self,dados):
        self.base=dados
        self.x_treinamento, self.x_teste, self.y_treinamento, self.y_teste = train_test_split(self.base.Temperatura,self.base.Total,test_size = 0.3,random_state = 0)
        self.x_treinamento = pd.DataFrame(self.x_treinamento)
        self.y_treinamento = pd.DataFrame(self.y_treinamento)
        self.treino = self.x_treinamento.merge(self.y_treinamento,on = self.x_treinamento.index)

        self.x_teste = pd.DataFrame(self.x_teste)
        self.y_teste = pd.DataFrame(self.y_teste)
        self.teste = self.x_teste.merge(self.y_teste,on = self.x_teste.index)
        self.teste = self.teste[(self.teste.Temperatura > self.treino.Temperatura.min())&(self.treino.Temperatura < self.treino.Temperatura.max())]
        
    def get_treino(self):
        return self.treino
    
    def get_teste(self):
        return self.teste
```

```python
#Separação de dados treino/teste
dados_previsao = treino_teste(dados_filtrados)
dados_treino = dados_previsao.get_treino()
dados_teste = dados_previsao.get_teste()
```

```python
#Previsão dos dados de Teste

modelo_previsao = previsores(dados_treino)
previsao_interpolacao = modelo_previsao.prev_interpl(dados_teste.Temperatura)
previsao_linear = modelo_previsao.prev_linear(dados_teste.Temperatura)
```

```python
#Inserir em Dataset avaliação

dados_teste["Previsao_Interpolacao"] = previsao_interpolacao
dados_teste["Previsao_Linear"] = previsao_linear
```

```python
#Grafico de dados

plt.plot(dados_teste.Total,"o",previsao_interpolacao,"-",previsao_linear,"-")
```

```python
dados_teste["Erro_Absoluto_Interpolacao"] = 1
dados_teste["Erro_Absoluto_Interpolacao"] = dados_teste.apply(lambda x:(round(abs(dados_teste.Total-dados_teste.Previsao_Interpolacao)/dados_teste.Total,2)))
```

```python
dados_teste["Erro_Absoluto_LR"] = 1
dados_teste["Erro_Absoluto_LR"] = dados_teste.apply(lambda x:(round(abs(dados_teste.Total-dados_teste.Previsao_Linear)/dados_teste.Total,2)))
```
